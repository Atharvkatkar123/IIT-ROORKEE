```python
x = df[['Balls','Strike_Rate','Dot_Percentage']].values
y = df['Runs'].values
n = len(x)
features= x.shape[1]

iteration = 2500
learning_rate = 0.00001
m = np.zeros(features)
c = 0

# For MSE
curve = []

for i in range(iteration):

    random_index = np.random.randint(n)  
    xi = x[random_index]
    yi = y[random_index]

    y_predict = np.dot(m,xi) + c
    
    m1 = -2*xi*(yi-y_predict)
    m_old = m


    c1 = -2*(yi-y_predict)
    c_old = c


    m_s_e1 = (yi - y_predict)**2
    curve.append(m_s_e1)

    m = m-learning_rate*m1
    c = c-learning_rate*c1

    #print(f"Mse is: {m_s_e1}")
    
    if len(curve) > 1 and abs(curve[-2] - curve[-1]) < 1e-6:
        print(f"Final m: {m} \n Final c : {c}")
        break

print(f"Final Parameters: m = {m}, c = {c}")
print(f"Final predictions: {m*x + c}")

x_new = []
balls = int(input("Enter balls"))
x_new.append(balls)
SR = float(input("Enter Strike Rate"))
x_new.append(SR)
Dot = float(input("Enter dot %"))
x_new.append(Dot)

x_new = np.array(x_new)
y_pred = np.dot(m, x_new) + c
print(f"Predicted Runs: {y_pred}")
```
