```python
height = x.shape[0]
height
width = x.shape[1]
width

mid_h = height//2
mid_w = width//2

t_left = x[:mid_h,:mid_w]
t_right = x[:mid_h,mid_w:]
b_left = x[mid_h:,:mid_w]
b_right = x[mid_h:,mid_w:]
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Load the image
image = Image.open('gray_man.jpg')  # Replace with your image path
image_array = np.array(image, dtype=np.float32)

# Square the pixel values (darkens the image)
squared_image = np.square(image_array)

# Normalize squared image to the range 0-255
squared_min, squared_max = squared_image.min(), squared_image.max()
normalized_squared_image = 255 * (squared_image - squared_min) / (squared_max - squared_min)

# Brighten alternative: Square root transformation
brightened_image = np.sqrt(image_array)

# Normalize brightened image to the range 0-255
brightened_min, brightened_max = brightened_image.min(), brightened_image.max()
normalized_brightened_image = 255 * (brightened_image - brightened_min) / (brightened_max - brightened_min)

# Convert back to uint8
normalized_squared_image = normalized_squared_image.astype(np.uint8)
normalized_brightened_image = normalized_brightened_image.astype(np.uint8)

# Plot the results
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_array.astype(np.uint8))
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(normalized_squared_image)
plt.title("Squared (Darkened) Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(normalized_brightened_image)
plt.title("Square Root (Brightened) Image")
plt.axis('off')

plt.show()

```
