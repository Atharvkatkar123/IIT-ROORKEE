```python
import numpy as np


x = np.array([10, 11, 12])
y = np.array([13, 14, 19])


steps = float(input("Enter steps: "))


def m_range(start, stop, step):
    current = start
    while current <= stop:
        yield current
        current += step

def c_range(start, stop, step):
    current = start
    while current <= stop:
        yield current
        current += step

best_m, best_c = 0, 0
min_error = float('inf')


for m in m_range(-100 * steps, 100 * steps, steps):
    for c in c_range(-500 * steps, 100 * steps, steps):

        y_predicted = m * x + c
        

        error = np.sum((y - y_predicted) ** 2)
        

        if error < min_error:
            min_error = error
            best_m, best_c = m, c

print(f"Best m: {best_m}, Best c: {best_c}, Min Error: {min_error}")
```

```python
import numpy as np

x = np.array([10, 11, 12])
y = np.array([13, 14, 19])

m = np.random.rand() 
c = np.random.rand()
steps = 0.01
cycles = 1000

for cycle in range(cycles):

    y_pred = m * x + c
    dm = -2 * np.sum((y - y_pred) * x) / len(x)
    dc = -2 * np.sum(y - y_pred) / len(x)
    m -= steps * dm
    c -= steps * dc

print(f"Best m: {m}, Best c: {c}")
```
