```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler

def prepare_embeddings(triplets, model_name, model, max_triplets=5):
    labels, embeddings = [], []

    for i, (anchor, pos, neg) in enumerate(triplets[:max_triplets]):
        if model_name == "fasttext":
            get_vec = lambda x: np.mean([model.get_word_vector(w) for w in x.split()], axis=0)
        elif model_name == "use":
            vecs = model([anchor, pos, neg])
            get_vec = lambda x: vecs[["anchor", "positive", "negative"].index(x)].numpy()
        else:  # SBERT
            vecs = model.encode([anchor, pos, neg])
            get_vec = lambda x: vecs[["anchor", "positive", "negative"].index(x)]

        # Store embeddings and labels
        for tag, text in zip(["anchor", "positive", "negative"], [anchor, pos, neg]):
            emb = get_vec(tag)
            embeddings.append(emb)
            labels.append(f"{text} ({tag}) - #{i+1}")

    return np.array(embeddings), labels

def plot_2d_embeddings(embeddings, labels, model_name):
    # Standardize before PCA for better spread
    scaler = StandardScaler()
    scaled_embeddings = scaler.fit_transform(embeddings)

    # Reduce to 2D
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(scaled_embeddings)

    # Scatter Plot
    fig, ax = plt.subplots(figsize=(10, 6))
    colors = {"anchor": "blue", "positive": "green", "negative": "red"}

    for i, (x, y) in enumerate(reduced):
        tag = labels[i].split("(")[1].split(")")[0]
        ax.scatter(x, y, c=colors[tag], label=tag if tag not in labels[:i] else "", s=80, edgecolor='k')
        ax.text(x + 0.01, y + 0.01, labels[i].split(" (")[0], fontsize=8)

    ax.set_title(f"2D Embedding Map ({model_name.upper()}) - Euclidean Geometry")
    ax.legend(loc='best')
    ax.grid(True)
    plt.tight_layout()
    plt.show()

# Run for each model
for model_name, model_obj in [("sbert", model), ("use", model2), ("fasttext", ft3)]:
    embs, lbls = prepare_embeddings(triplets, model_name, model_obj)
    plot_2d_embeddings(embs, lbls, model_name)
```

```python
import matplotlib.pyplot as plt

def triplet_loss(anchor, positive, negative, margin=0.2):
    d_ap = cosine_distances([anchor], [positive])[0][0]
    d_an = cosine_distances([anchor], [negative])[0][0]
    return max(d_ap - d_an + margin, 0)

def encode_triplet_losses_per_example(model_name, model, num_examples=21):
    losses = []
    for anchor, pos, neg in triplets[:num_examples]:
        if model_name == "fasttext":
            get_vec = lambda x: np.mean([model.get_word_vector(w) for w in x.split()], axis=0)
        elif model_name == "use":
            vecs = model([anchor, pos, neg])
            get_vec = lambda x: vecs[["anchor", "positive", "negative"].index(x)].numpy()
        else:  # SBERT
            vecs = model.encode([anchor, pos, neg])
            get_vec = lambda x: vecs[["anchor", "positive", "negative"].index(x)]

        a_vec = get_vec("anchor")
        p_vec = get_vec("positive")
        n_vec = get_vec("negative")
        loss = triplet_loss(a_vec, p_vec, n_vec)
        losses.append(loss)
    return losses

# Compute per-example triplet losses for first 3 triplets
sbert_losses = encode_triplet_losses_per_example("sbert", model)
use_losses = encode_triplet_losses_per_example("use", model2)
fasttext_losses = encode_triplet_losses_per_example("fasttext", ft3)

# Triplet labels (optional, clearer x-axis)
triplet_labels = [
    f"{a} vs {p} vs {n}"
    for a, p, n in triplets[:21]
]

# Plotting
def plot_triplet_losses(triplet_labels, sbert_losses, use_losses, fasttext_losses):
    x = np.arange(len(triplet_labels))
    width = 0.25

    fig, ax = plt.subplots(figsize=(12, 6))
    ax.bar(x - width, sbert_losses, width, label='SBERT', color='skyblue')
    ax.bar(x, use_losses, width, label='USE', color='lightgreen')
    ax.bar(x + width, fasttext_losses, width, label='FastText', color='salmon')

    ax.set_xticks(x)
    ax.set_xticklabels(triplet_labels, rotation=20, ha='right', fontsize=9)
    ax.set_xlabel("Triplets")
    ax.set_ylabel("Triplet Loss")
    ax.set_title("Triplet Loss Comparison Examples")
    ax.legend()
    ax.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Show the plot
plot_triplet_losses(triplet_labels, sbert_losses, use_losses, fasttext_losses)
```
